This notebook uses raw recruiting data to count signed recruits of each star rating from 2006-2014 by team and year.

```python
import pandas as pd
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())

df = pd.read_csv("~/projects/Graduate Thesis/Data/Commits Data/commits.csv", header = 0)

df.head()
```

```python
data = df[df["year"] >= 2006]
data2 = data[data["year"] < 2015]
signed = data2[data2["status"] == 'signed']

signed.head()
```

```python
query = """

WITH stars AS (

SELECT
  year,
  team,
  CASE WHEN stars <= 2 OR stars IS NULL THEN 1 ELSE 0 END AS "2_star",
  CASE WHEN stars = 3 THEN 1 ELSE 0 END AS "3_star",
  CASE WHEN stars =4 THEN 1 ELSE 0 END AS "4_star",
  CASE WHEN stars =5 THEN 1 ELSE 0 END AS "5_star"
FROM signed

 )

SELECT 
  year,
  team,
  SUM("2_star") AS "num_2_stars_committed",
  SUM("3_star") AS "num_3_stars_committed",
  SUM("4_star") AS "num_4_stars_committed",
  SUM("5_star") AS "num_5_stars_committed"
FROM stars
GROUP BY 
  team, 
  year
  
"""

finalData = pysqldf(query)

finalData.head()
```

```python
finalData.to_csv("~/projects/Graduate Thesis/Data/Commits Data/AggregatedRecruits06-14.csv")
```
